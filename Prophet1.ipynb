```python
import pandas as pd
import matplotlib.pyplot as plt
from prophet import Prophet
from prophet.plot import plot_plotly

```

```python
data=pd.read_csv("C:\\Users\\27797\\OneDrive\\Documents\\SalesDataTable.csv")
```

```python
data.head()
```

```python
#display columns
data.columns
```

```python
data=data.rename(columns={'Date':'ds','Units Sold':'y'})
```

```python
#display first five rows
data.head()
```

```python
#change ds to datetime format
data['ds'] = pd.to_datetime(data['ds'])
```

```python
data.head()
```

```python
#initialise model and fit the model
model = Prophet()
model.fit(data)
```

```python
#make forecast for the next 30 days
future=model.make_future_dataframe(periods=30)
#make predictions based on the forecast
forecast=model.predict(future)
```

```python
#plot or visualise the forecast
model.plot(forecast)
```
